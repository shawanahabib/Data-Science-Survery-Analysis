```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# import and view data
# view data
file_path = 'https://cocl.us/datascience_survey_data'
df = pd.read_csv(file_path)
df
```

```python
# set index to data science topics
df = pd.read_csv(file_path, index_col='Unnamed: 0')
df
```

```python
# import matplotlib to plot bar chart
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# sort dataframe into desceinding order by column 'Very Interested'
df_sort = df.sort_values(by='Very interested', ascending = False, axis=0)
df_sort.head()
```

```python
# convert value to percentage using: value = value*100/number of respondents
# round to 2 decimal places
df_sort = round(df_sort*100/2233,2)
df_sort.head()
```

```python
# plot the bar chart
ax = df_sort.plot(kind='bar', figsize=(20,8), width=0.8, color=('#5cb85c','#5bc0de', '#d9534f'),fontsize = 14)
ax.set_title('Percentage of Respondents'' Interest in Data Science Areas', fontsize=16)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.set_visible(False)

# set individual bar lables using above list
for i in ax.patches:
    width, height = i.get_width(), i.get_height()
    x, y = i.get_xy()
    ax.annotate('{}%'.format(height), (x, y + height + 1), fontsize=14)
```
